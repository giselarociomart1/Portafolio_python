#  Desafío - Transformación y manipulación de datos (Parte II)

1. Carga los datos de cada pestaña del archivo US_Regional_Sales_Data.xlsx en un
DataFrame independiente. (hint: Puedes obtener los nombres de las hojas usando
pd.ExcelFile(archivo).sheet_names).

```python
# Primero importamos la librería Pandas.
import pandas as pd
```

```python
# Cargamos las hojas del archivo.
archivo = 'US_Regional_Sales_Data.xlsx'

# Creamos ExcelFile para ver los nombres de las hojas.
excel = pd.ExcelFile(archivo)
print(excel.sheet_names)  
```

```python
# Cargamos cada hoja en un DataFrame.
df_sales_orders=pd.read_excel(archivo, sheet_name='Sales Orders Sheet')
df_customers=pd.read_excel(archivo, sheet_name='Customers Sheet')
df_store_locations=pd.read_excel(archivo, sheet_name='Store Locations Sheet')
df_products=pd.read_excel(archivo, sheet_name='Products Sheet')
df_sales_team=pd.read_excel(archivo, sheet_name='Sales Team Sheet')
```

2. Cruza todos los DataFrames usando validación many_to_one, y guarda el resultado
en una nueva variable llamada df_base.

```python
# Vemos los nombres de las columnas de los dataframes.
df_sales_orders.columns
```

```python
df_customers.columns
```

```python
df_store_locations.columns
```

```python
df_products.columns
```

```python
df_sales_team.columns
```

```python
# Cruzamos todos los DataFrames usando validación many_to_one.
df_base = (
    df_sales_orders
    .merge(df_customers, on='_CustomerID', how='left', validate='many_to_one')
    .merge(df_store_locations, on='_StoreID', how='left', validate='many_to_one')
    .merge(df_products, on='_ProductID', how='left', validate='many_to_one')
    .merge(df_sales_team, on='_SalesTeamID', how='left', validate='many_to_one')
)

# Visualizamos el dataframe resultante.
df_base
```

3. Agrega las siguientes columnas a df_ base:
    ProcurementDays=OrderDate-ProcuredDate
    ShippingDays=ShipDate-OrderDate
    DeliveryDays=DeliveryDate-ShipDate
    CustomerDays=ShippingDays+DeliveryDays
   hint: Asegúrate de que las columnas sean de tipo datetime.

```python
# Nos aseguramos de que las columnas de fechas sean tipo datetime.
df_base['ProcuredDate'] = pd.to_datetime(df_base['ProcuredDate'])
df_base['OrderDate'] = pd.to_datetime(df_base['OrderDate'])
df_base['ShipDate'] = pd.to_datetime(df_base['ShipDate'])
df_base['DeliveryDate'] = pd.to_datetime(df_base['DeliveryDate'])

# Creamos las nuevas columnas.
df_base['ProcurementDays'] = (df_base['OrderDate'] - df_base['ProcuredDate']).dt.days
df_base['ShippingDays'] = (df_base['ShipDate'] - df_base['OrderDate']).dt.days
df_base['DeliveryDays'] = (df_base['DeliveryDate'] - df_base['ShipDate']).dt.days
df_base['CustomerDays'] = df_base['ShippingDays'] + df_base['DeliveryDays']

# Mostamos el dataframe con las nuevas columnas.
df_base
```

4. Agrega una nueva columna discreta al df_base, con el nombre CustomerDaysInterval
que clasifique los valores de la columna CustomerDays en los siguientes intervalos:
○ 0 to 15 days
○ 15 to 30 days
○ 30 to 45 days
○ 45 to 60 days
○ 60 to 75 days
○ 75 to 90 days
hint: Utiliza pd.cut con estas etiquetas de Intervalo. Puedes utilizar la propiedad
`.dt.days` para convertir la columna CustomerDays a valores numéricos. La
propiedad `.dt.days` te permite obtener directamente la representación numérica de
la diferencia en días a partir de un objeto DateInterval.

```python
# Definimos límites de los intervalos.
bins = [0, 15, 30, 45, 60, 75, 90]

# Definimos las etiquetas para cada rango.
etiquetas = ['0 to 15 days','15 to 30 days','30 to 45 days','45 to 60 days','60 to 75 days','75 to 90 days']

# Creamos la nueva columna.
df_base['CustomerDaysInterval'] = pd.cut(df_base['CustomerDays'], bins=bins,labels= etiquetas, right=True)

# Visualizamos el dataframe.
df_base
```

5. Utiliza una tabla pivote para mostrar el conteo de órdenes agrupando por Sales
Team en las filas y CustomerDaysInterval en las columnas. Este reporte debe
cuadrar con los datos de la pestaña Reporte1 del archivo reportes.xlsx.

```python
# Creamos la tabla pivote (Reporte1).
pivot_reporte1 = pd.pivot_table(
    df_base,
    index='Sales Team',                 
    columns='CustomerDaysInterval',     
    values='OrderNumber',               
    aggfunc='count',                   
    fill_value=0,
    margins = True
)

# Mostramos el resultado.
pivot_reporte1

# Exportamos a excel para comparar resultados.
ruta = r"C:...\Desafios\reporte1.xlsx"  #ruta en Windows

with pd.ExcelWriter(ruta, mode='w', engine='openpyxl') as writer:
    pivot_reporte1.to_excel(writer, sheet_name='Reporte1', index=True)

print(f"Reporte1 exportado correctamente en: {ruta}")
```

6. Agrega la siguiente columna calculada a la base GrossMargin = OrderQuantity*(UnitPrice*(1-Discount Applied)-UnitCost).Visión de Negocio: El margen bruto es un indicador financiero que muestra la diferencia entre
los ingresos generados por la venta de un producto o servicio y los costos directos asociados
con su producción o adquisición. 

```python
# Creamos la columna calculada GrossMargin.
df_base['GrossMargin'] = (df_base['Order Quantity'] * (df_base['Unit Price'] * (1 - df_base['Discount Applied']) - df_base['Unit Cost']))

# Mostramos el resultado.
df_base
```

7. En esta parte debes aplicar un porcentaje a la columna GrossMargin, pero ese
porcentaje dependerá del intervalo en que se encuentre GrossMargin. CommissionsPercentage =  % de comisión según la tabla de comisiones. hint: Utiliza pd.cut con labels numéricas para que la columna resultante sea de tipo float.

```python
# Asignamos el porcentaje de comisión según el intervalo de GrossMargin.

# Primero definimos intervalos y etiquetas.
bins = [0, 100, 1000, 10000, 100000]
etiquetas = [0.05, 0.10, 0.15, 0.20] 

# Ahora creamos la columna que categoriza los los datos de la columna GrossMargin. También nos aseguramos que sea de tipo float.
df_base['CommissionsPercentage'] = pd.cut(df_base['GrossMargin'], bins=bins, labels=etiquetas, right=True).astype(float)

# Mostramos el dataframe.
df_base
```

8. Calcula el monto de la comisión, utilizando la Tabla del Anexo. CommissionsAmount = GrossMargin*CommissionsPercentage.

```python
# Calculamos el monto de la comisión a partir de las columnas GrossMargin y CommissionsPercentage.
df_base['CommissionsAmount'] = df_base['GrossMargin'] * df_base['CommissionsPercentage']

# Mostamos el dataframe. 
df_base
```

9. Calcula la comisión sobre el margen bruto. NetMargin = GrossMargin-CommissionsAmount.

```python
# Calculamos el margen neto usando las columnas GrossMargin y CommissionsAmount.
df_base['NetMargin'] = df_base['GrossMargin'] - df_base['CommissionsAmount']

# Mostamos el dataframe. 
df_base
```

10. Utilizando una tabla pivote, muestra la suma de los valores GrossMargin, NetMargin y
CommissionAmount agrupando por Sales Team en las filas. Este reporte debe cuadrar con
los datos de la pestaña Reporte2 del archivo reportes.xlsx.

```python
# Creamos la tabla pivote (Reporte2).
pivot_reporte2 = pd.pivot_table(
    df_base,
    index='Sales Team',
    values=['GrossMargin', 'NetMargin', 'CommissionsAmount'],
    aggfunc='sum',
    fill_value=0,
    margins = True
)

# Mostramos el resultado.
pivot_reporte2

# Exportamos a excel para comparar resultados.
ruta = r"C:...\Desafios\reporte2.xlsx"  #ruta en Windows

with pd.ExcelWriter(ruta, mode='w', engine='openpyxl') as writer:
    pivot_reporte2.to_excel(writer, sheet_name='Reporte2', index=True)

print(f"Reporte2 exportado correctamente en: {ruta}")
```
